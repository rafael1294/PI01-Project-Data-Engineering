```python
import pandas as pd
import numpy as np
```

```python
# Importamos la data de las plataformas presentadas en la Datasets y agregamos la columna Id_Plataform para identificar 
# cada plataforma
data1 = pd.read_csv("./Datasets/amazon_prime_titles-score.csv",encoding="utf-8")
data1.insert(loc=1, column="id_plataform", value='a')

data2 = pd.read_csv("./Datasets/disney_plus_titles-score.csv",encoding="utf-8")
data2.insert(loc=1, column="id_plataform", value='d')

data3 = pd.read_csv("./Datasets/hulu_titles-score (2).csv",encoding="utf-8")
data3.insert(loc=1, column="id_plataform", value='h')

data4 = pd.read_csv("./Datasets/netflix_titles-score.csv",encoding="utf-8")
data4.insert(loc=1, column="id_plataform", value='n')
```

```python
# Unimos la data en una sola tabla
df=pd.concat([data1,data2,data3,data4],ignore_index=True)
df.head(2)
```

```python
# Generamos la columna "id"
id=df["id_plataform"]+df["show_id"]
id.head(2)
```

```python
# Agregamos el campo "id" al dataframe
df.insert(2,"id",id.values)
df.head(5)
```

```python
df.info()
```

```python
# Eliminamos los campos que no seran de utilidad para las consignas solicitadas
df.drop(columns=["director","cast","country","description"], axis = 1, inplace=True)
```

```python
# Campos de texto a minúsculas
df['type']=df['type'].str.lower()
df['title']=df['title'].str.lower()
df['listed_in']=df['listed_in'].str.lower()
```

```python
# Cambiamos el formato del campo "date_added" a AAAA-mm-dd
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
# Comprobamos el cambio de formato para date_added
df.head()
```

```python
# Reemplazamos los valores vacios de rating por "G"
df['rating']=df['rating'].fillna("g")
```

```python
# Buscamos duplicados en el campo "title"
df['title'][df['title'].duplicated(keep=False)]
```

```python
# Verificamos algunos duplicados
df1=df['title']=='young adult'
df[df1]
```

```python
# Generamos una nueva tabla "id_title" para hacer únicos los títulos por tipo y año de estreno
df['id_title']=df['release_year'].apply(str)+"-"+df['title']+"-"+df['type']
```

```python
# Reemplazamos los valores nan
df.replace(np.nan,"",inplace=True)
```

```python
# Contabilizamos los datos vacios
df[df==""].count()
```

```python
df.head()
```

```python
# Reemplazamos los datos vacios en duration por algun duplicado con el mismo "id_title"
for i in df[df["duration"]==""].index: #df[df["duration"]==""].index nos da una lista con indices de vacio para la columna duration
    for j in df[df["id_title"].duplicated(False)].index:#df[df["id_title"].duplicated(False)].index nos da una lista con indices duplicados sin excluir los primeros encontrados.
        if i!=j and  df.loc[j,"duration"]!="" and df.loc[i,"id_title"]==df.loc[j,"id_title"]:
            df.loc[i,"duration"]=df.loc[j,"duration"]
            break
```

```python
# Contabilizamos los datos vacios y verificamos que se pudo disminuir los datos vacios de 482 a 451
df[df==""].count()
```

```python
# Separamos el campo "duration" en "duration_int" y "duration_type"
df[['duration_int','duration_type']]=df.duration.str.split(' ',expand=True)
```

```python
# Comprobamos la separación del campo "duration"
df.head(5)
```

```python
# Generamos la tabla dftitle
dftitle=df.loc[:,("id_plataform","type","title","release_year","rating","duration","score")]
dftitle.drop_duplicates(inplace=True)
dftitle.reset_index(inplace=True,drop=True)
dftitle.info()
```

```python
# Creamos los archivos csv de la data 
dftitle.to_csv("Tablas/dftitle.csv")
```
